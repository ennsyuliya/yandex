```python
import pandas as pd
```

### 1

Объедините самостоятельно таблицы subcategory_dict и data. 

Результат сохраните в переменной data_subcategory и выведите на экран первые 10 строк новой таблицы.

Подсказка

В качестве аргументов укажите следующие значения: subcategory_dict, on='subcategory_id' и how='left'.

```python
data = pd.read_excel('./seo_data.xlsx', sheet_name='traffic_data')
subcategory_dict = pd.read_excel('./seo_data.xlsx', sheet_name='subcategory_ids')
```

```python
data_subcategory = data.merge(subcategory_dict, on='subcategory_id', how='left')
data_subcategory.head(10)
```

### 2


Объедините таблицы data_subcategory и category_dict. 

Результат сохраните в переменной data_final и выведите на экран первые 10 строк новой таблицы.

```python
category_dict = pd.read_excel('./seo_data.xlsx', sheet_name='category_ids')
```

```python
data_final = data_subcategory.merge(category_dict, on='category_id', how='left')
data_final
```
